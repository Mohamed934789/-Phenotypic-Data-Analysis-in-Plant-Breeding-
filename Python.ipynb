```python
sns.set_theme(style="whitegrid")
sugar_cols = ["Sucrose","Glucose","Fructose"]
df["TotalSugar"] = df[sugar_cols].sum(axis=1, skipna=True)
pairs = [
    ("BIO","Sweight"),
    ("LA_avg","BIO"),
    ("SLA_avg","BIO"),
    ("TotalSugar","Sweight"),
    ("Sucrose","Sweight")
]

for x,y in pairs:
    plt.figure(figsize=(7,5))  # حجم البلوت
    sns.regplot(
        data=df,
        x=x, 
        y=y,
        scatter_kws={"alpha":0.6, "s":60},   # شفافية وحجم النقاط
        line_kws={"color":"black", "lw":2},  # خط الانحدار
        color='green'
    )
    
    plt.title(f"{y} vs {x}", fontsize=14, fontweight="bold")
    plt.xlabel(x, fontsize=12)
    plt.ylabel(y, fontsize=12)
    plt.tight_layout()  # ترتيب البلوت
    
    plt.show()
    plt.savefig("3.png")
```

```python
traits = ["BIO","Sweight","LA_avg","SLA_avg","TotalSugar"]
parent_means = df_parent.groupby("Var")[["LL1","LL2","LW1","LW2","LA1","LA2","LDW1","LDW2","SLA1","SLA2","BIO","Sweight"]].mean()


parent_means["LA_avg"]  = parent_means[["LA1","LA2"]].mean(axis=1)
parent_means["SLA_avg"] = parent_means[["SLA1","SLA2"]].mean(axis=1)


hi = parent_means.reindex(["Xena","H94"]).mean(axis=0) 

hi_trait = parent_means.reindex(["Xena","H94"])[["BIO","Sweight","LA_avg","SLA_avg"]].max()

# علِّم الـ RILs المتفوّقة
for t in ["BIO","Sweight","LA_avg","SLA_avg"]:
    df[f"Transgressive_{t}"] = df[t] > float(hi_trait[t])


elite = df[df["Transgressive_BIO"] | df["Transgressive_Sweight"]][["RIL No.","BIO","Sweight","LA_avg","SLA_avg"]].sort_values("Sweight", ascending=False)
elite.head(15)
```

```python
def top_n_bar(df, trait, n=10):
    top = df.nlargest(n, trait)
    plt.figure(figsize=(10,4))
    sns.barplot(data=top, x='RIL No.', y=trait)
    plt.title(f"Top {n} RILs by {trait}")
    plt.xticks(rotation=45)
    plt.tight_layout(); plt.show()

top_n_bar(df, 'Sweight', 10)
top_n_bar(df, 'BIO', 10)
top_n_bar(df, 'ExtnFruAr', 10)
top_n_bar(df, 'StCaExtnVol', 10)
top_n_bar(df, 'StCaExtMas', 10)
top_n_bar(df, 'TotalSugar', 10)
plt.savefig("9.png")
```
